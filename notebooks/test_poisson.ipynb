```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def gain(x):
    return 3500*(sigmoid(x/1000)-0.5)
```

```python
# draw poisson counts for mean = m
m = 1000
n = 1000
counts = np.random.poisson(m, n)
```

```python
counts.var()/counts.mean()
```

```python
rates = np.logspace(1, 3.5, 100)
ratios = []
for rate in rates:
    counts = np.random.poisson(rate, n)
    counts = gain(counts)
    ratio = counts.var()/counts.mean()
    ratios.append(ratio)
ratios = np.array(ratios)
```
